```python
import pandas as pd
import numpy as np
from skorch import NeuralNetClassifier
import torch
import torch.nn as nn

from skorch import NeuralNetClassifier
from skorch.callbacks import EarlyStopping
import torch.nn.functional as F 

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
df = pd.read_csv('ADANIENT.csv')
```

```python
df['target'] = np.where(df['Adj Close'] - df['Adj Close'].shift()>0,1,0)
```

```python
X_train = df[['Open','High','Low']].values
y_train = df['target'].values
```

```python
X_train = torch.tensor(X_train)
y_train = torch.tensor(y_train)

X_train = X_train.to(dtype=torch.float32) 
y_train = y_train.to(torch.int64) 
```

```python
# Define your PyTorch model
class StockClassifier(nn.Module):
    def __init__(self, input_dim, num_classes):
        super(StockClassifier, self).__init__()
        self.fc1 = nn.Linear(3, 64)
        self.fc2 = nn.Linear(64, 2)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

    # Create an instance of your custom neural network
mymodel = StockClassifier(1,1).to(device)

# Wrap your PyTorch model with Skorch
net = NeuralNetClassifier(
    module=mymodel,
    module__input_dim=X_train.shape[1],
    module__num_classes=2,
    criterion=nn.CrossEntropyLoss,
    optimizer=torch.optim.Adam,
    max_epochs=1000,  # Number of training epochs
    #callbacks=[EarlyStopping(patience=10)],  # Optional early stopping
)

# Train the model
net.fit(X_train, y_train)
```

```python
# Evaluate the model
accuracy = net.score(X_train, y_train)
print(f"Test Accuracy: {accuracy}")
```
