```python
import re

# Sample data in a dictionary format (your actual data may vary)
data = {
    'item_weight': "KALAS ClassictMo Natural-lodized Sea salt NET WEIGHT:14.11oZ-400g",
    'item_weight_2': "ORIGINAL Nosalt Alternative NETWT.110Z.311 g",
    'height': "40A39x39c15,4x15.4in/po 2PLEPASSEURSLAGEBCAPASFOLHAS 9old Renova Renova",
    'width': "96 ct. PARTY ESSENTIALS' 9.875 in x9.875 in 25x25cm",
    'depth': "9.8 inch/25 cm",
}

# Regular expression patterns for weights and dimensions
patterns = {
    'weight': r'(\d+\.?\d*)\s*(oz|lb|g|kg)',
    'height': r'(\d+\.?\d*)\s*(in|cm)',
    'width': r'(\d+\.?\d*)\s*(in|cm)',
    'depth': r'(\d+\.?\d*)\s*(in|cm)',
}

# Function to extract values based on entity name
def extract_entity_value(text, entity_name):
    if 'weight' in entity_name:
        match = re.search(patterns['weight'], text, re.IGNORECASE)
    elif 'height' in entity_name:
        match = re.search(patterns['height'], text, re.IGNORECASE)
    elif 'width' in entity_name:
        match = re.search(patterns['width'], text, re.IGNORECASE)
    elif 'depth' in entity_name:
        match = re.search(patterns['depth'], text, re.IGNORECASE)
    else:
        match = None
    
    if match:
        return match.group(0)  # Return the full match
    else:
        return "No value found"

# Extract and display entity values
for entity_name, text in data.items():
    value = extract_entity_value(text, entity_name)
    print(f"{entity_name}: {value}")
```

```python
import os
import shutil
from tqdm import tqdm

def copy_jpg_images(src_folder, dest_folder):
    # Ensure the destination folder exists
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)
    
    # List all files in the source directory
    files = os.listdir(src_folder)
    
    # Filter out .jpg files
#     jpg_files = [f for f in files if f.lower().endswith('.jpg')]
    
    # Copy each .jpg file to the destination directory
    for jpg_file in tqdm(jpg_files):
        src_path = os.path.join(src_folder, jpg_file)
        dest_path = os.path.join(dest_folder, jpg_file)
        shutil.copy2(src_path, dest_path)  # copy2 preserves metadata
#         print(f'Copied {jpg_file} to {dest_folder}')

# Define the source and destination directories
source_directory = r'D:\Images_ml\train'
destination_directory = r'D:\Images_ml\test_1000each'

# Call the function
copy_jpg_images(source_directory, destination_directory)
```

```python
import numpy as np

# Define a function to check if two bounding boxes overlap
def boxes_overlap(box1, box2):
    # box = [x1, y1, x2, y2] where (x1, y1) is top-left and (x2, y2) is bottom-right
    x1_1, y1_1, x2_1, y2_1 = box1
    x1_2, y1_2, x2_2, y2_2 = box2

    # Check if boxes overlap
    horizontal_overlap = (x1_1 < x2_2) and (x2_1 > x1_2)
    vertical_overlap = (y1_1 < y2_2) and (y2_1 > y1_2)
    
    return horizontal_overlap and vertical_overlap

# Example bounding boxes
bounding_boxes = [
    [347.05970, 612.42520, 420.7693, 564.469540],
    [615.38684, 863.60390, 541.67725, 911.5595],
    # Add the rest of your bounding boxes here
]

# Check for overlap between all pairs of bounding boxes
for i in range(len(bounding_boxes)):
    for j in range(i+1, len(bounding_boxes)):
        if boxes_overlap(bounding_boxes[i], bounding_boxes[j]):
            print(f"Bounding box {i} overlaps with bounding box {j}")
```

```python
def boxes_same_line(box1, box2, threshold=10):
    _, y1_1, _, y2_1 = box1
    _, y1_2, _, y2_2 = box2
    
    # Check if both boxes are nearly on the same horizontal line (allowing some tolerance)
    return abs(y1_1 - y1_2) < threshold and abs(y2_1 - y2_2) < threshold

# Example to check if boxes are in the same line
for i in range(len(bounding_boxes)):
    for j in range(i+1, len(bounding_boxes)):
        if boxes_same_line(bounding_boxes[i], bounding_boxes[j]):
            print(f"Bounding box {i} is in the same line as bounding box {j}")
```
