```python
# Q1
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# define factorial
x = sp.symbols('x')
factorial = sp.factorial(x)

# define Stirling's approximation
stirlings_approx = sp.sqrt(2 * sp.pi * x) * (x / sp.exp(1))**x

# Calculate the limit
limit_ratio = sp.limit(stirlings_approx / factorial, x, sp.oo)

print("Limit of g(x)/f(x) as x approaches infinity:", limit_ratio)
```

```python
# Q2
f = lambda x: [sp.factorial(i) for i in x]
g = sp.lambdify(x, stirlings_approx, 'numpy')

# Define x values for plotting
x_values = np.linspace(0, 5, 100)

# Plot f(x) and g(x)
plt.plot(x_values, f(x_values), label='Factorial function f(x)',color='r')
plt.plot(x_values, g(x_values), label="Stirling's approximation g(x)",color='g')
plt.xlabel('x')
plt.ylabel('Value')
plt.title("Factorial function vs Stirling's approximation")
plt.legend()
plt.show()
```
