```python
import pandas as pd
import math

# Read the CSV file into a pandas DataFrame
# df = pd.read_csv('marks.csv')

# Function to split marks into 12 columns as specified
def split_marks(marks):
    marks_distribution = [min(marks, 4) for _ in range(10)]  # First 10 columns have maximum 4 marks each
    remaining_marks = marks - sum(marks_distribution)
    marks_distribution.extend([min(remaining_marks/2, 30), min(remaining_marks/2, 30)])  # Last 2 columns have maximum 30 marks each
    return marks_distribution

# Apply the split_marks function to each row to create new columns
df[['Marks_1', 'Marks_2', 'Marks_3', 'Marks_4', 'Marks_5', 'Marks_6', 'Marks_7', 'Marks_8', 'Marks_9', 'Marks_10', 'Marks_11', 'Marks_12']] = df.apply(lambda row: pd.Series(split_marks(row['Marks'])), axis=1)

# Drop the original 'marks' column
# df.drop('marks', axis=1, inplace=True)

# Write the transformed DataFrame to a new CSV file
# df.to_csv('transformed_marks.csv', index=False)
```

```python
def split_number(number):
    part_45 = round(number * 0.45 * 2) / 2  # 45% part in multiples of 0.5
    part_55 = round(number * 0.55 * 2) / 2  # 55% part in multiples of 0.5
    return part_45, part_55
```
