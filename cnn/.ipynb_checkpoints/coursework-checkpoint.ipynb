```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.nn.functional as F
import my_utils as mu
import matplotlib.pyplot as plt
import numpy as np

# functions to show an image
```

```python
transform = transforms.Compose([
    # Randomly resize and crop to 32x32, scaling the content between 80% to 120%
    transforms.RandomResizedCrop(size=(32, 32), scale=(0.8, 1.2)),

    # Randomly flip the image horizontally with a probability of 50%
    transforms.RandomHorizontalFlip(),

    # Randomly rotate the image by up to 15 degrees
    transforms.RandomRotation(degrees=15),

    # Randomly adjust the brightness, contrast, saturation, and hue
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),

    # Convert the image to a tensor
    transforms.ToTensor(),

    # Normalize the image (mean and std dev for each channel)
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

batch_size = 128

trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=8)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=8)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
# class VGG_block(nn.Module):
#     def __init__(self, num_convs, input_channels, output_channels):
#         super(VGG_block, self).__init__()
#         self.num_convs = num_convs
#         self.convs = nn.ModuleList()
#         self.spatial_pools = nn.ModuleList()
#         self.linears = nn.ModuleList()
        
#         for i in range(num_convs):
#             # Convolutional layer
#             self.convs.append(nn.Conv2d(input_channels, output_channels, kernel_size=3, padding=1))
            
#             # Layers for computing a_k
#             self.spatial_pools.append(nn.AdaptiveAvgPool2d((1, 1)))
#             self.linears.append(nn.Linear(output_channels, 1))
            
#             # Update input channels for the next convolutional layer
#             input_channels = output_channels
        
#         self.relu = nn.ReLU()
#         self.max_pool = nn.MaxPool2d(kernel_size=2, stride=2)

#     def forward(self, x):
#         conv_outputs = []
#         scalar_weights = []

#         # Pass through each conv layer and calculate a_k
#         for i in range(self.num_convs):
#             conv_out = self.convs[i](x)
#             conv_out = self.relu(conv_out)
#             conv_outputs.append(conv_out)
            
#             # Compute a_k
#             # I made a change to x
#             pooled = self.spatial_pools[i](conv_out)  # [batch_size, out_channels, 1, 1]
#             pooled = torch.flatten(pooled, 1)  # [batch_size, out_channels]
#             a_k = self.linears[i](pooled)  # [batch_size, 1]
#             a_k = self.relu(a_k)  # Apply ReLU to a_k for non-linearity
#             scalar_weights.append(a_k)
        
#         # Combine the outputs: O = sum(a_k * conv_k)
#         out = torch.zeros_like(conv_outputs[0])
#         for i in range(self.num_convs):
#             a_k = scalar_weights[i].view(-1, 1, 1, 1)  # Reshape a_k to broadcast
#             out += a_k * conv_outputs[i]  # Multiply and accumulate
        
#         out = self.max_pool(out)
#         return out
import torch
import torch.nn as nn

class VGG_block(nn.Module):
    def __init__(self, num_convs, input_channels, output_channels):
        super(VGG_block, self).__init__()
        self.num_convs = num_convs
        self.convs = nn.ModuleList()
        self.spatial_pools = nn.ModuleList()
        self.linears = nn.ModuleList()
        
        for i in range(num_convs):
            # Convolutional layer
            self.convs.append(nn.Conv2d(input_channels, output_channels, kernel_size=3, padding=1))
            
            # Layers for computing a_k (using the original input x, not conv_out)
            self.spatial_pools.append(nn.AdaptiveAvgPool2d((1, 1)))
            self.linears.append(nn.Linear(input_channels, 1))  # Note: Linear layer now uses input_channels
            
            # Update input channels for the next convolutional layer
            input_channels = output_channels
        
        self.relu = nn.ReLU()
        self.max_pool = nn.MaxPool2d(kernel_size=2, stride=2)

    def forward(self, x):
        conv_outputs = []
        scalar_weights = []

        # Path 1: Compute a_k using the original input x
        for i in range(self.num_convs):
            pooled = self.spatial_pools[i](x)  # [batch_size, input_channels, 1, 1]
            pooled = torch.flatten(pooled, 1)  # [batch_size, input_channels]
            a_k = self.linears[i](pooled)  # [batch_size, 1]
            a_k = self.relu(a_k)  # Apply ReLU to a_k for non-linearity
            scalar_weights.append(a_k)
            
            # Path 2: Pass through the convolutional layer
            conv_out = self.convs[i](x)
            conv_out = self.relu(conv_out)
            conv_outputs.append(conv_out)
            
            # Update x to be the output of the current convolution for the next layer
            x = conv_out
        
        # Combine the outputs: O = sum(a_k * conv_k)
        out = torch.zeros_like(conv_outputs[0])
        for i in range(self.num_convs):
            a_k = scalar_weights[i].view(-1, 1, 1, 1)  # Reshape a_k to broadcast
            out += a_k * conv_outputs[i]  # Multiply and accumulate
        
        out = self.max_pool(out)
        return out
```

```python
class VGG(nn.Module):
    def __init__(self, conv_arch):
        super(VGG, self).__init__()
        in_channels = 3
        self.conv_arch = conv_arch
        for i, (num_convs, out_channels) in enumerate(conv_arch):
            self.add_module('vgg_block{0}'.format(i), VGG_block(num_convs, in_channels, out_channels))
            in_channels = out_channels
        
        # Spatial average pooling to reduce to [batch_size, out_channels, 1, 1]
        self.global_avg_pool = nn.AdaptiveAvgPool2d((1, 1))

        self.last = nn.Sequential(nn.Flatten(), nn.Linear(out_channels, 4096),
                                  nn.ReLU(), nn.Dropout(0.5), nn.Linear(4096, 4096),
                                  nn.ReLU(), nn.Dropout(0.5), nn.Linear(4096, 10))

    def forward(self, x):
        out = x
        for i in range(len(self.conv_arch)):
            out = self._modules['vgg_block{0}'.format(i)](out)
        out = self.global_avg_pool(out)  # Shape becomes [batch_size, out_channels, 1, 1]
        out = self.last(out)
        return out
```

```python
conv_arch = ((1, 64), (1, 128), (1, 256), (1, 512), (1, 512))
# Determine if a GPU is available and set the device accordingly

net = VGG(conv_arch)  # Move the model to the GPU if available


loss = nn.CrossEntropyLoss()
lr, wd = 0.1, 0.001
# Creare your optimizer here. Use SGD with weight decay wd and learning rate lr.
optimizer = torch.optim.Adam(net.parameters(), lr=lr, weight_decay=wd)
```

```python
num_epochs = 5 # learning rate 0.1
mu.train_ch3(net, trainloader, testloader, loss, num_epochs, optimizer)
```
