```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, f1_score,precision_score
from sklearn.linear_model import LogisticRegression

import warnings
warnings.filterwarnings("ignore")

df = pd.read_excel('Data_08_02_2024.xlsx')

# drop correlated features with correlation of more than 0.95
corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
print('column dropped due to high correlation', to_drop)

# Drop features 
df.drop(to_drop, axis=1, inplace=True)

threshold = len(df) * 0.25 # Calculate 20% of the total number of rows
dropped_columns = list(df.columns[df.isna().sum() > threshold])
df = df.drop(columns=dropped_columns, axis=1)
print("Columns deleted:", dropped_columns)

df.drop(columns='business_id',inplace=True)

y = df[['hard_closure' ,'soft_closure']]
X = df.drop(columns=['hard_closure' ,'soft_closure'])

X['per_capita_income'] = X['per_capita_income'].str.replace('[$,]', '').astype(float)

object_cols = X.select_dtypes(include=['object']).columns
float_cols = X.select_dtypes(exclude=['object']).columns

object_imputer = SimpleImputer(strategy='most_frequent')
X[object_cols] = object_imputer.fit_transform(X[object_cols])

float_imputer = SimpleImputer(strategy='mean')
X[float_cols] = float_imputer.fit_transform(X[float_cols])

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoder.fit(X[object_cols])
X_cat_encoded = encoder.transform(X[object_cols])

# Concatenate the encoded features with the numeric features
X_encoded = pd.DataFrame(X_cat_encoded.toarray(), columns=encoder.get_feature_names_out(object_cols))
X_encoded[float_cols] = X[float_cols]

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.20, random_state=42)

# balance the data
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_hard, y_train_hard = smote.fit_resample(X_train, y_train['hard_closure'])

X_train_soft, y_train_soft = smote.fit_resample(X_train, y_train['soft_closure'])

# logistics regression

Hard_model = LogisticRegression()
Hard_model.fit(X_train_hard, y_train_hard)

plt.figure(figsize=(20, 6))
plt.bar(X_train_hard.columns, Hard_model.coef_[0])

plt.xticks(rotation=90)

plt.xlabel('Feature')
plt.ylabel('Significance')
plt.title('hard closure:Significance of Features')

Soft_model = LogisticRegression()
Soft_model.fit(X_train_soft, y_train_soft)

plt.figure(figsize=(20, 6))
plt.bar(X_train_soft.columns, Soft_model.coef_[0])

plt.xticks(rotation=90)

plt.xlabel('Feature')
plt.ylabel('Significance')
plt.title('soft closure:Significance of Features')

# Prediction

y_pred_hard = Hard_model.predict(X_test)
y_pred_soft = Soft_model.predict(X_test)

print('***hard closure***',classification_report(y_test['hard_closure'],y_pred_hard),sep='\n')

print('****soft closure****',classification_report(y_test['soft_closure'],y_pred_hard),sep='\n')

print('*****hard_closure prediction******')
print(pd.DataFrame({'True':y_test['hard_closure'],'Pred':y_pred_hard}))

print('*****soft_closure prediction******')
print(pd.DataFrame({'True':y_test['soft_closure'],'Pred':y_pred_hard}))
```
