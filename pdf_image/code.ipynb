```python
# STEP 1 
# pip install PyMuPDF Pillow
# code adjusted as per https://stackoverflow.com/questions/75321414/python-page-object-has-no-attribute-getimagelist-where-i-try-to-extract-im

# import libraries 
import fitz 
import io 
from PIL import Image 

# STEP 2 
# file path you want to extract images from 
file = "நெஞ்சை_உருக்கும்_நீதிக்கதைகள்.pdf"

# open the file 
pdf_file = fitz.open(file) 

# STEP 3 
# iterate over PDF pages 
for page_index in range(len(pdf_file)): 

    # get the page itself 
    page = pdf_file[page_index] 
    image_list = page.get_images() 

    # printing number of images found in this page 
    if image_list: 
        print(f"[+] Found a total of {len(image_list)} images in page {page_index}") 
    else: 
        print("[!] No images found on page", page_index) 
    for image_index, img in enumerate(page.get_images(), start=1): 

        # get the XREF of the image 
        xref = img[0] 

        # extract the image bytes 
        base_image = pdf_file.extract_image(xref) 
        image_bytes = base_image["image"] 

        x = is_text_only(image_bytes)
        if x==False:
            print(page_index,'Image')
        # get the image extension 
        image_ext = base_image["ext"] 
        
        image_name = f"page{page_index + 1}_image{image_index}.{image_ext}"

        # save the image 
        image = Image.open(io.BytesIO(image_bytes)) 
        image.save('images/'+image_name) 

        print(f"Image {image_index} on page {page_index + 1} saved as {image_name}")
```

```python
import cv2
import numpy as np

def is_text_only(image_bytes):
    # Convert image bytes to numpy array
    np_img = np.frombuffer(image_bytes, np.uint8)
    # Decode image
    img = cv2.imdecode(np_img, cv2.IMREAD_GRAYSCALE)
    
    # Calculate edges using Canny edge detector
    edges = cv2.Canny(img, 100, 200)
    
    # Calculate the percentage of edges
    edge_density = np.sum(edges) / (img.shape[0] * img.shape[1])
    
    # Set a threshold for distinguishing text-only vs text+figure
    threshold = 0.02  # Adjust this value based on your dataset
    
    if edge_density < threshold:
        return True  # Likely a text-only image
    else:
        return False
```
