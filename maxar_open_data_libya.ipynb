```python
m = leafmap.Map(height="600px")
m.add_basemap("SATELLITE")
m.add_raster("libya_sample.tif", layer_name="Image")
m.add_layer_manager()
m
```

```python
sam_kwargs = {
    "points_per_side": 32,
    "pred_iou_thresh": 0.86,
    "stability_score_thresh": 0.92,
    "crop_n_layers": 1,
    "crop_n_points_downscale_factor": 2,
    "min_mask_region_area": 80,
}
```

```python
sam = SamGeo(
    model_type="vit_h",
    sam_kwargs=sam_kwargs,
)
```

```python
sam.generate("/Users/poojapandey/Documents/libya_trial.tif", output="mask_libya.tif", foreground=True)
```

```python
raster_to_vector("mask_libya.tif", output="mask_libya.shp")
```

```python
sam.show_masks(cmap="binary_r")
```

```python
sam.show_anns(axis="off", opacity=1, output="annotation_libya.tif")
```

```python
leafmap.image_comparison(
    "/Users/poojapandey/Documents/libya_trial.tif",
    "annotation.tif",
    label1="Image",
    label2="Segmentation",
)
```

```python
overlay_images("/Users/poojapandey/Documents/ibya_trial.tif", "annotation_libya.tif", backend="TkAgg")
```
